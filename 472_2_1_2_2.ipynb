```python
import json
import numpy as np
import pandas as pd

from sklearn import preprocessing
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

data_file = open('goemotions.json')
data = json.load(data_file)
```

```python
#Collecting the words and their frequencies
tokens = []


for lists in data:
    for text in lists[:1]:
        tokens.append(text)
        
cv = CountVectorizer()
        
c_arr = cv.fit(tokens)
print("Vocabulary: ", cv.vocabulary_["that"])
#Prints out the vocabulary, commenting it out because it takes up the whole page on github
```

```python
#Printing the size of the vocabulary
print(len(cv.vocabulary_))
```

```python
#Converting Json array to Numpy Array
npData = np.array(data)
cv = CountVectorizer()
labEncoder = preprocessing.LabelEncoder()
npDataTokens = cv.fit_transform(npData[:,0])
npDataEmotions = labEncoder.fit_transform(npData[:,1])
npDataSentiments = labEncoder.fit_transform(npData[:,2])
```

```python
#Splitting the data in two halves
train_tokens, test_tokens = train_test_split(npDataTokens, test_size=0.2, train_size=0.8, shuffle=False)
train_emotions, test_emotions = train_test_split(npDataEmotions, test_size=0.2, train_size=0.8, shuffle=False)
train_sentiments, test_sentiments = train_test_split(npDataSentiments, test_size=0.2, train_size=0.8, shuffle=False)
```

```python
#Emotions 2.3.3 MLP
from sklearn.neural_network import MLPClassifier
```

```python
emotions_clf = MLPClassifier(max_iter=40)
emotions_clf.fit(train_tokens, train_emotions)
```

```python
sentiments_clf = MLPClassifier()
sentiments_clf.fit(train_tokens, train_sentiments)
```

```python
#Gridsearch
import joblib
```

```python
filename1 = "sentiments_Base_MLP_model.sav"
joblib.dump(sentiments_clf,open(filename1,'wb'))
```

```python
filename2 = "emotions_Base_MLP_model.sav"
joblib.dump(emotions_clf,open(filename2,'wb'))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
mlp_emotions_parameters = {
    'solver': ["adam", "sgd"],
    'hidden_layer_sizes' : [(10,5),(15,10)],
    'activation' : ["sigmoid", "relu", "tanh", "identity"]
}
top_mlp_emotions = GridSearchCV(estimator=MLPClassifier(max_iter=2), param_grid=mlp_emotions_parameters)
```

```python
top_mlp_emotions.fit(train_tokens, train_emotions)
print("done")
```

```python
filename3 = "emotions_Top_MLP_model.sav"
joblib.dump(top_mlp_emotions,open(filename3,'wb'))
```

```python
mlp_sentiments_parameters = {
    'solver': ["adam", "sgd"],
    'hidden_layer_sizes' : [(10,5),(15,10)],
    'activation' : ["sigmoid", "relu", "tanh", "identity"]
}
top_mlp_sentiments = GridSearchCV(estimator=MLPClassifier(max_iter=2), param_grid=mlp_sentiments_parameters)
```

```python
top_mlp_sentiments.fit(train_tokens, train_sentiments)
print("done")
```

```python
filename4 = "sentiments_Top_MLP_model.sav"
joblib.dump(top_mlp_sentiments,open(filename4,'wb'))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
emotions_base_mlp_results = emotions_clf.predict(test_tokens)
print(confusion_matrix(test_emotions, emotions_base_mlp_results))
```

```python
print(classification_report(test_emotions, emotions_base_mlp_results))
```

```python
sentiments_results = sentiments_clf.predict(test_tokens)
print(confusion_matrix(test_sentiments, sentiments_results))
```

```python
print(classification_report(test_sentiments, sentiments_results))
```

```python
top_mlp_emotions_results = top_mlp_emotions.predict(test_tokens)
print(confusion_matrix(test_emotions,top_mlp_emotions_results))
```

```python
print(classification_report(test_emotions,top_mlp_emotions_results))
```

```python
top_mlp_sentiments_results = top_mlp_sentiments.predict(test_tokens)
```

```python
print(confusion_matrix(test_sentiments, top_mlp_sentiments_results))
```

```python
print(classification_report(test_sentiments, top_mlp_sentiments_results))
```
