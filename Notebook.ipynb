```python
import json
  
data_file = open('goemotions.json')
data = json.load(data_file)
```

```python
import matplotlib.pyplot as plt
```

```python
emotionsList = {}
sentimentList = {}
for cell in data:
    emotion = cell[1]
    sentiment = cell[2]
    if (emotion not in emotionsList):
        emotionsList[emotion] = 1
    else:
        emotionsList[emotion] += 1
        
        
    if (sentiment not in sentimentList):
        sentimentList[sentiment] = 0
    else:
        sentimentList[sentiment] += 1
    
```

```python
print(emotionsList)
print(sentimentList)
```

```python
sentiments = list(sentimentList.keys())
sentimentAmount = list(sentimentList.values())
plt.title("Distribution of the sentiments")
plt.bar(range(len(sentimentList)),sentimentAmount,tick_label=sentiments)
plt.savefig("Sentiment_Distribution", format='png')
plt.show()
```

```python
emotions = list(emotionsList.keys())
emotionsAmount = list(emotionsList.values())
plt.title("Distribution of Emotions")
plt.pie(emotionsAmount,labels=emotions,radius=1.1,rotatelabels=True,startangle=-10)
plt.savefig("Emotions_Distribution", format='png')
plt.show()
```
